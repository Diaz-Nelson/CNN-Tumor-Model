```python
import numpy as np
from tensorflow.keras.models import load_model, save_model
import cv2
import matplotlib.pyplot as plt
```

```python
labels = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
```

```python
model = load_model("D:\MasterProgramming\WIP_2025\CODE\PROJECTS\BrainTumorLearning\\tumor_model.keras")
```

```python
def preprocess_image(image_path, target_size):
    img = cv2.imread(image_path)
    img = cv2.resize(img,(target_size,target_size))
    return [img]
```

```python
def predict_given_image(model,image_array,labels):
    predictions = model.predict(image_array)
    predicted_index = np.argmax(predictions)
    predicted_label = labels[predicted_index]
    confidence_score = predictions[0][predicted_index]
    return predicted_label,confidence_score

```

```python
img_path = "D:\MasterProgramming\WIP_2025\CODE\PROJECTS\BrainTumorLearning\m.jpg"
```

```python
X = np.array(preprocess_image(img_path,150))
```

```python
print(len(X))
```

```python
predicted_label, score = predict_given_image(model,X,labels)
```

```python
plt.imshow(X[0])
plt.title(f"Predicted: {predicted_label} with a confidence score of {score*100:.3f}%")
```
